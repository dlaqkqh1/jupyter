```python
%matplotlib inline

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/state-population.csv')
```

```python
data.shape
```

```python
data.info()
```

```python
data['population'] = data['population'].fillna(data['population'].median())
```

```python
data = data[data['state/region']!='USA']
```

```python
data[(data['ages']==
    'under18') & (data['year'] == 2012)].groupby(['state/region', 'ages']).mean()[['population']].plot(kind='bar', figsize=(25, 15));
```

```python
data[(data['ages']==
    'total') & (data['year'] == 2012)].groupby(['state/region', 'ages']).mean()[['population']].plot(kind='bar', figsize=(25, 15));
```
