```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = "http://www.naver.com"
res=req.get(url)
```

```python
print(res.text)
```

```python
url = "https://search.naver.com/search.naver"
res = req.get(url,params={'query':'정보처리기사'})
res.text
```

```python
#헤더 사용은 일반적인 크롤링이 안될때 사용하는 방법
url = "https://search.naver.com/search.naver"
headers ={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36'}
res = req.get(url,headers = headers,params={'query':'정보처리기사'})
res.text
```

```python
# soup = bs(응답받은 데이터 , 해석프로그램)
# 해석프로그램 : lxml,html.parser,html5lib - 속도순 lxml 이 가장 빠름.
soup = bs(res.text,'html.parser')

#soup.find(태그명, 아이디 또는 클래스) -> 하나만선택 가능
soup.find('a', class_='lnk_tit').text
```

```python
# 여러개 선택 가능 find_all 은 리스트 형이므로 인덱스 후 .text 해야 볼수 있음
soup.find_all('a',class_='lnk_tit')[0].text 
```

```python
for i in soup.find_all('a',class_='lnk_tit'):
    print(i.text)
```

## 검색 뉴스 데이터 가져오기

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = "https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=100&oid=056&aid=0010688671"
res = req.get(url)
soup = bs(res.text,"html.parser")


newsTitle1 = soup.find('h3',id='articleTitle').text
newsTitle2 = soup.select('div.article_info > h3')[0].text #선택자로 가져오고 싶을때 여러개라 리스트형 find_all 과 같이 리스트 형이다.
#newsTitle2 = soup.select_one('div.article_info > h3').text #선택자형식이지만 하나만 가져오고 싶을때 _one 를 사용한다.뒤에도 인덱스 지정x 
newsContents = soup.find('div',id='articleBodyContents').text

print('제목:',newsTitle2)
print('내용:',newsContents.strip())   
```

## 네이버 실시간 검색어 가져오기

```python
url = "http://www.naver.com/"
res = req.get(url)
soup = bs(res.text,"html.parser")

keywords = soup.findAll('span',class_='ah_k',limit=20)

for index in range(len(keywords)):
    print(str(index+1) + "." + keywords[index].text)
```

## 네이버 음악 top10 가져오기

```python
url = "https://music.naver.com/listen/top100.nhn?domain=TOTAL"
res = req.get(url)
soup = bs(res.text,"html.parser")
```

```python
rank_list = soup.select('td.ranking')
rank_list
```

```python
title_list = soup.select('a._title > span.ellipsis')
```

```python
artist_list = soup.select('a._artist > span.ellipsis')
artist_list
```

```python
len(rank_list),len(title_list),len(artist_list)
```

```python
ranks = []
titles = []
artists = []
for index in range(11):
    ranks.append(rank_list[index].text)
    titles.append(title_list[index].text)
    artists.append(artist_list[index].text.replace('\n','').replace('\t','').replace('\r','').strip())
```
