```python
import requests
import pymysql
from bs4 import BeautifulSoup as bs
response = requests.get('https://music.bugs.co.kr/chart')
html = response.text
soup = bs(html, 'html.parser')
```

```python
soup
```

```python
soup.find_all('p', class_='title')
soup.find_all('p', class_='artist')
```

```python
music_list = []
for title, artist in zip(soup.find_all('p', class_='title'), soup.find_all('p', class_='artist')[0]):
    print(title.text, artist.text) 
    music_list.append({'title':title.text.strip(), 'artist':artist.text.strip()}) # 양쪽에 '\n' 이 들어 있으므로 strip() 함수로 제거
```

```python
music_list
```

```python
conn = pymysql.connect(host='localhost', user='root', password='cksdn12z', database='jupyter')
#charset='utf-8'
```

```python
curs = conn.cursor(pymysql.cursors.DictCursor)
curs
```

```python
sql ="create table music (song_id int primary key auto_increment, title varchar(100), artist varchar(50))"
curs.execute(sql)
```

```python
sql = "insert into music(title, artist) values (%s, %s)"
```

```python
for music in music_list:
    curs.execute(sql, (music['title'], music['artist']))
```

```python
curs.execute('select * from music')
```
