```python
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
print(movies.shape)
movies.head()
```

```python
movies_df = movies[['id', 'title', 'genres', 'vote_average', 'vote_count']]
movies_df
```

```python
movies_df[['genres']]
```

```python
from ast import literal_eval
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['genres'] = movies_df['genres'].apply(lambda x : [y['name'] for y in x])
```

```python
movies_df[['genres']][:1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
movies_df['genre_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))
```

```python
movies_df[['genre_literal']][:1]
```

```python
count_vect = CountVectorizer(min_df = 0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies_df['genre_literal'])
print(genre_mat.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim)
```

```python
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:1])
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    
    title_movie=df[df['title']==title_name]
    title_index=title_movie.index.values
    similar_indexes=sorted_ind[title_index, :(top_n)]
    
    print(similar_indexes)
    similar_indexes=similar_indexes.reshape(-1)
   
    return df.iloc[similar_indexes]
```

```python
similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'McFarland, USA', 10)
similar_movies[['title','vote_average']]
```

```python
movies_df[['title', 'vote_average', 'vote_count']].sort_values('vote_average', ascending=False)[:10]
```

```python
C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6)

def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ((v/(v+m))*R) + ((m/(m+v))*C)

movies_df['weighted_vote']=movies.apply(weighted_vote_average, axis=1)
```

```python
movies_df[['title', 'vote_average', 'vote_count', 'weighted_vote']].sort_values('weighted_vote', ascending = False)[:10]
```

```python
def find_sim_movie2(df, sorted_ind, title_name, top_n=10):
    
    title_movie=df[df['title']==title_name]
    title_index=title_movie.index.values
    similar_indexes=sorted_ind[title_index, :(top_n)]
    
    print(similar_indexes)
    similar_indexes=similar_indexes.reshape(-1)
    similar_index = similar_indexes[similar_indexes!=title_index]
   
    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]
```

```python
similar_movies = find_sim_movie2(movies_df, genre_sim_sorted_ind, 'McFarland, USA', 10)
similar_movies[['title','vote_average']]
```
